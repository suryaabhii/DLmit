```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generating a simple dataset
np.random.seed(0)
X = np.random.rand(100, 1)
y = 3 * X.squeeze() + 2 + np.random.randn(100) * 0.1  # y = 3x + 2 + noise

# Adding a constant for intercept in statsmodels
X_with_intercept = sm.add_constant(X)

# Simple Linear Regression
model = sm.OLS(y, X_with_intercept)
results = model.fit()

# Summary of regression
print(results.summary())

# Plotting the regression line
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, results.predict(X_with_intercept), color="red", label="Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
